```python
import requests
import pandas as pd
from api_keys import geoapify_key

# Set parameters to search for supermarkets in a specific area
categories = 'commercial.supermarket'
filter_param = 'place:512bfb51b070235ec059b304e6cb3d1a4340f00101f901b60c060000000000c0020992031043616c61766572617320436f756e7479'

params = {
    'categories': categories,
    'filter': filter_param,
    'apiKey': geoapify_key
}

# Set base URL
base_url = "https://api.geoapify.com/v2/places"

# Make API request
response = requests.get(base_url, params=params)

# Convert response to JSON format
data = response.json()

# Create empty lists to store supermarket data
names = []
lats = []
longs = []

# Iterate through the features in the JSON data
for feature in data['features']:
    
    # Extract name, latitude, and longitude of the supermarket
    name = feature['properties']['name']
    lat = feature['geometry']['coordinates'][1]
    long = feature['geometry']['coordinates'][0]
#   address = feature['properties']['formatted']
#   city = feature['properties']['cities']
    
    # Append data to the respective lists
    names.append(name)
    lats.append(lat)
    longs.append(long)
#    addresses.append(address)
#    cities.append(city)

# Create pandas DataFrame from the data
supermarket_calavera = pd.DataFrame({'Name': names, 'Latitude': lats, 'Longitude': longs})

# Print the DataFrame
supermarket_calavera
```

```python
import hvplot.pandas
import geopandas as gpd
import requests
import plotly.express as px

# Create a scattermapbox plot with Plotly Express
fig = px.scatter_mapbox(
    supermarket_calavera, 
    lat="Latitude", 
    lon="Longitude", 
    hover_name="Name",
#     hover_data=["Address", "City"],
    zoom=8.5,
    size_max=10
)

# Set the map style
fig.update_layout(mapbox_style="open-street-map")
fig.update_traces(marker=dict(size=12))
                              
# Add a title to the plot
fig.update_layout(title="Supermarkets in Calaveras County, CA")

# Show the plot
fig.show()
```
